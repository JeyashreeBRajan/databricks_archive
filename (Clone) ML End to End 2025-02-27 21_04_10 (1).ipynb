### Part 3 - Training data with sklearn

```python
from sklearn.model_selection import train_test_split

X = data.drop(['quality'],axis=1)
Y = data.quality

#Split the training dataset (60%)
X_train, X_rem, y_train, Y_rem = train_test_split(X, Y, train_size=0.6, random_state=142)

#Split the remaining dataset for test and validation (40%)
X_val, X_test, Y_val, y_test = train_test_split(X_rem, Y_rem, train_size=0.5, random_state=142)
```

### Part 4 - Built baseline model

```python
#1. mlflow
import mlflow
import mlflow.pyfunc
import mlflow.sklearn
from mlflow.models import infer_signature
#from mlflow.utils.environment import _mlflow_conda_env

#2. sklearn
import sklearn
from sklearn.ensemble import RandomForestClassifier 
from sklearn.metrics import roc_auc_score

#3. Numpy/Pandas
import numpy as np

#4. Other
import cloudpickle
import time
```

```python
class SKlearnModeWrapper(mlflow.pyfunc.PythonModel):
    def __init__(self, model):
        self.model = model
    def predict(self, context, model_input):
        return self.model.predict_proba(model_input)[:,1]

with mlflow.start_run(run_name="Untuned_random_forest_wine-quality-prediction"):
    #Set model parameter
    n_estimators= 100
    max_depth = 5
    max_features = 3

    #Create and train model
    rf = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth, max_features=max_features)
    rf.fit(X_train, y_train)

    #Use the model to make prediction on test data
    #prediction = rf.predict(X_test)
    predictions_test = rf.predict_proba(X_test)[:,1]

    #Log the model parameters for this run
    mlflow.log_param("n_estimators", n_estimators)
    mlflow.log_param("max_depth", max_depth)
    mlflow.log_param("max_features", max_features)

    #Define a metric to evaluate a model - Area Under Curve
    #mse = mean_squared_error(y_test, prediction)
    auc_score = roc_auc_score(y_test, predictions_test)

    #Log the metric from this run
    mlflow.log_metric("roc_auc_score", auc_score)
    wrappedModel= SKlearnModeWrapper(rf)

    #Log the model created by this run
    #Log the model with Signature - In MLflow, a model signature precisely defines the schema for model inputs, outputs, and any additional parameters required for effective model operation. 
    signature = infer_signature(X_train, wrappedModel.predict(None, X_train))

    #Log the model created by this run
    mlflow.pyfunc.log_model("random-forest-model", python_model=wrappedModel, signature=signature)


```

### Part 6 Experiment with new model 
Try creating models with different libraries and hyperparameters and find the best model. Don't just get satisfied with one score.

```python
#Above model was build with RandomForestClassifier from sklearn. Now we are going to try with Hyperopt and XGBoost

#hyperopt
from hyperopt import fmin, tpe, hp, SparkTrials, Trials, STATUS_OK, space_eval
from hyperopt.pyll import scope

#xgboost
import xgboost as xgb
import mlflow.xgboost

#others
from math import exp
import numpy as np
```

```python
#We are going to try substituting different parameters on Permutation and combination method

search_space = {
   'max_depth': scope.int(hp.quniform('max_depth', 3, 10, 1)),
   'learning_rate': hp.loguniform('learning_rate', -3, 0),
   'reg_alpha': hp.loguniform('reg_alpha', -6, -1),
   'reg_lambda': hp.loguniform('reg_lambda', -6, -1),
   'min_child_weight': hp.loguniform('min_child_weight', -6, 0),
   'objective': 'binary:logistic',
   'seed': 123
}
```

```python
def train_mode(params):
    #Enable autologging for XGBoost
    mlflow.xgboost.autolog()
    with mlflow.start_run(nested=True):
        train = xgb.DMatrix(X_train, label=y_train)
        validation = xgb.DMatrix(X_val, label=Y_val)
        booster = xgb.train(params, train, num_boost_round=100, evals=[(validation, 'validation')])
        validation_predictions = booster.predict(validation)
        auc_score = roc_auc_score(Y_val, validation_predictions)
        mlflow.log_metric('auc_score', auc_score)

        signature = infer_signature(X_train, validation_predictions)
        mlflow.xgboost.log_model(booster, 'model', signature=signature)
        #set loss=-1*auc_score to minimize loss
        return {'status': STATUS_OK, 'loss': -auc_score, 'booster': booster}
```

```python
tpe
```

```python
algo = tpe.suggest
```

Enable parallelism

```python
#Since we are going to run many models, enable parallelism
#Greater parallelism will increase the speed of the run, but a less optimal hyperparameter search will be performed.
#A reasonable value for parallelism is square root of max_evals
from hyperopt import SparkTrials
spark_trials = SparkTrials(parallelism=5)
```

Define fmin function

```python
#We have defined all the parameters used in fmin above. Now we are going to run the model
#max_evals =10 -> so, 10 models will be created
with mlflow.start_run(run_name='XGBoost_models_wine_quality'):
    best_params = fmin(
      fn=train_mode, 
      space=search_space, 
      algo=algo, 
      max_evals=10, 
      trials=spark_trials)
```

```python
#Display best performing model
best_run = mlflow.search_runs(order_by=['metrics.auc_score DESC']).iloc[0]
print(f"Best run: {best_run.run_id} with auc_score: {best_run['metrics.auc_score']}")
```

```python
#Register the new model with best score
new_model_version = mlflow.register_model(f"runs:/{best_run.run_id}/model", model_name)

time.sleep(10)
```

```python
#Archive the old version
client.transition_model_version_stage(model_name, new_model_version.version, "Archived")
#Promote the new version to Production
client.transition_model_version_stage(model_name, new_model_version.version, "Production")
```

```python
model = mlflow.pyfunc.load_model(f"models:/{model_name}/Production")
print(f"AUC: {roc_auc_score(y_test, model.predict(X_test))}")
```
